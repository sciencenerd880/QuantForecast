```python
import joblib
import pandas as pd
```

```python
# Load the saved model from the .pkl file
loaded_model = joblib.load('../weights/best_rf_model.pkl')
loaded_model
```

```python
df = pd.read_csv("../data/X_test.csv")
df.head()
```

```python
df.columns
```

```python
feature_set = ['num_candles','price_range_pct', 'num_res_hits', 'num_sup_hits', 'num_res_breakouts',
       'num_sup_breakouts', 'alternation', 'last_candle_return',
       'last_candle_pct_above_support',
       'concentration_score_scaled', 'alternation_score_scaled',
       'pre_sideway_5_bar_change', 'pre_sideway_10_bar_change']
target = ["post_sideway_5_bar_change"]
```

```python
# Features (independent variables)
X_test = df[feature_set]

# Target (dependent variable)
y_test = df[target]
```

```python
# Use the loaded model to make predictions
y_pred = loaded_model.predict(X_test)
```

```python
y_pred*100
```

```python
# Add the predicted values to the original DataFrame as a new column
df['predicted_post_sideway_5_bar_change'] = y_pred
```

```python
# Save the updated DataFrame to a CSV file
df.to_csv('../data/test_predictions.csv', index=False)

print("New output predictions from test data saved as 'test_predictions.csv'")
```
